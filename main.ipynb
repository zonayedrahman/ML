```python
# Scatter and density plots
def plotScatterMatrix(df):
    x = df1.iloc[:, 1:]
    # output
    y = df1.iloc[:, 0]

    fig, axs = plt.subplots(56, 2, figsize=(18,70))

    col_names = df.columns[1:]
    for i in range(0,56):
        for j in range(0,2):
            axs[i, j].scatter(x.iloc[:,i*2+j],y, s=1)
            axs[i, j].set_xlabel(col_names[i*2+j])
            axs[i, j].set_ylim([-1,3])

    for ax in axs.flat:
        ax.set( ylabel='Class')

    fig.tight_layout()
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn import linear_model
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python

def basic_logistic_regression(df):
    x = df.iloc[:, 1:].values

    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()

    x = scale.fit_transform(x)

    poly = PolynomialFeatures(degree=1)
    x = poly.fit_transform(x)

    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.20, random_state=42)


    log_reg = linear_model.LogisticRegression(max_iter=3000)
    log_model = log_reg.fit(xtrain, ytrain)

    y_pred = log_model.predict(xtest)
    report = classification_report(ytest,y_pred)

    cm = confusion_matrix(ytest, y_pred)
    print (f"Confusion Matrix of {df.name}: \n", cm)
    # print(classification_report(ytest,y_pred))
    # print("Coefficients:",  log_model.coef_)
    # print("metrics:", )
    print ("Accuracy of training: ", log_model.score(xtrain, ytrain))
    print ("Accuracy : ", accuracy_score(ytest, y_pred))

    return report
```

```python
df1.name = "user_a"
df1_report = basic_logistic_regression(df1)
df2.name = "user_b"
df2_report = basic_logistic_regression(df2)
df3.name = "user_c"
df3_report = basic_logistic_regression(df3)
df4.name = "user_d"
df4_report = basic_logistic_regression(df4)
```

```python
def poly_logistic_regression(df, degree):
    x = df.iloc[:, 1:].values
    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()
    x = scale.fit_transform(x)

    poly = PolynomialFeatures(degree=degree)
    x = poly.fit_transform(x)

    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.20, random_state=42)



    log_reg = linear_model.LogisticRegression(max_iter=3000)
    log_model = log_reg.fit(xtrain, ytrain)

    y_pred = log_model.predict(xtest)

    cm = confusion_matrix(ytest, y_pred)
    print (f"Confusion Matrix of {df.name} with feature degree of {degree}: \n", cm)
    print("Coefficients:",  log_model.coef_)
    print ("Accuracy of training: ", log_model.score(xtrain, ytrain))
    print ("Accuracy : ", accuracy_score(ytest, y_pred))
    report = classification_report(ytest,y_pred)
    return report
```

```python
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score


def l1_k_fold_logistic_regression(df):
    x = df.iloc[:, 1:].values

    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()

    x = scale.fit_transform(x)


    poly = PolynomialFeatures(degree=2)
    x = poly.fit_transform(x)
    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.20, random_state=42)



    alphas = [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0, 100.0, 1000.0, 10000.0]
    # alphas = [1.0, 10.0, 100.0, 1000.0]
    best_alph = [0] * 9
    acc = [0] * 9


    scoring = {'accuracy' : make_scorer(accuracy_score),
               'precision' : make_scorer(precision_score),
               'recall' : make_scorer(recall_score),
               'f1_score' : make_scorer(f1_score)}


    print("L1 REGULARIZATION: ")
    for i in range(2, 11):
        cv = KFold(n_splits=i, random_state=42, shuffle=True)
        best_score = 0
        best_c = 0.0001
        report = None
        for c in alphas:
            log_reg = linear_model.LogisticRegression(penalty='l1',C=c,solver="liblinear", max_iter=3000)
            log_model = log_reg.fit(xtrain, ytrain)
            scores = cross_val_score(log_reg, x, y, scoring='accuracy', cv=cv, n_jobs=-1)

            # print(f"scores with fold={i} and c={c}:", max(scores))
            if max(scores) > best_score:
                best_score = max(scores)
                best_c = c

        print("best C:", best_c, "Highest acc: ", best_score)
        best_alph[i-2] = best_c
        acc[i-2] = best_score



    print("END:", best_alph, acc)
    return best_alph, acc



```

```python
from sklearn.svm import SVC, LinearSVC
from sklearn.metrics import classification_report, confusion_matrix


def svm_linl1(df, C):

    x = df.iloc[:, 1:].values

    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()

    x = scale.fit_transform(x)


    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.20, random_state=42)

    lsvc = LinearSVC(C=C, penalty="l1", dual=False, max_iter=20000, random_state=42)

    # fitting x samples and y classes
    lsvc.fit(xtrain, ytrain)


    score = lsvc.score(xtrain, ytrain)
    ypred = lsvc.predict(xtest)
    cm = confusion_matrix(ytest,ypred)

    report = classification_report(ytest,ypred)



    print(f"ACCURACY TRAINING SCORE OF {df.name} with C={C}:        ", lsvc.score(xtrain, ytrain) )
    print(f"ACCURACY SCORE OF {df.name} with C={C}:     ", accuracy_score(ytest, ypred) )

    # print(f"CONFUSION MATRIX OF {df.name}: \n", cm)
    #
    # print(f"REPORT ON LINEAR KERNAL WITH C={C}: \n", report)

    return cm, report

```

```python
def svm_poly(df, degree):
    x = df.iloc[:, 1:].values

    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()

    x = scale.fit_transform(x)


    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.20, random_state=42)



    # C = [0.0000001, 0.00001, 0.001, 0.01, 0.1, 1, 10, 100, 10000, 1000000]
    poly_svc = SVC(kernel="poly", degree=degree, random_state=42)

    # fitting x samples and y classes
    poly_svc.fit(xtrain, ytrain)


    score = poly_svc.score(xtrain, ytrain)
    ypred = poly_svc.predict(xtest)
    cm = confusion_matrix(ytest,ypred)

    report = classification_report(ytest,ypred)

    print(f"TRAINING SET ACCURACY SCORE OF {df.name} with poly kernal of degree={degree}:       ", poly_svc.score(xtrain, ytrain) )
    print(f"ACCURACY SCORE OF {df.name} with poly kernal of degree={degree}:        ", accuracy_score(ytest, ypred) )
    #
    # print(f"CONFUSION MATRIX OF {df.name}: \n", cm)
    #
    # print(f"REPORT ON LINEAR KERNAL WITH C={C}: \n", report)

    return cm, report
```

```python
def k_fold_svm_poly(df):


    x = df.iloc[:, 1:].values

    # output
    y = df.iloc[:, 0].values

    scale = StandardScaler()

    x = scale.fit_transform(x)


    xtrain, xtest, ytrain, ytest = train_test_split(
        x, y, test_size=0.20, random_state=42)

    best_degrees = [0] * 10
    acc = [0] * 10
    print(f"{df.name} poly kernal: ")
    for k in range(2, 11):
        cv = KFold(n_splits=k, random_state=42, shuffle=True)
        best_score = 0
        best_degree = 1
        for degree in range(1,11):
            poly_svc = SVC(kernel="poly", degree=degree, random_state=42)
            poly_svc.fit(xtrain, ytrain)
            scores = cross_val_score(poly_svc, x, y, scoring='accuracy', cv=cv, n_jobs=-1)
            print(f"For k={k} with degree={degree} scores:{scores}")
            if max(scores) > best_score:
                best_score = max(scores)
                best_degree = degree

        print("For K=", k, "best degree:", best_degree, "Highest acc: ", best_score)
        best_degrees[k-2] = best_degree
        acc[k-2] = best_score



    print("END:", best_degrees, acc)
    return best_degrees, acc
```
